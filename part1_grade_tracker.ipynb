```python
# raw students data given
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    # Clean the data remove whitespace
    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks = [int(m) for m in student["marks_str"].split(", ")]

    cleaned_students.append({"name": name, "roll": roll, "marks": marks})

    # Validate each name should contain only alphabets
    words = name.split()
    is_valid = all(word.isalpha() for word in words)
    validity = "✓ Valid name" if is_valid else "✗ Invalid name"

    # Print report card for each student
    print("=================================")
    print(f"Student : {name} — {validity}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("=================================")

# print roll number 103 in All caps and lower
for student in cleaned_students:
    if student["roll"] == 103:
        print(f"Roll 103 - ALL CAPS: {student['name'].upper()}")
        print(f"Roll 103 - lowercase: {student['name'].lower()}")
        break
```

```python
# given data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]


def get_grade(score):
    """Return grade based on marks."""
    if score >= 90:
        return "A+"
    elif score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "F"


# print each subject marks and grade using for loop

print(f"REPORT CARD - {student_name}")
print(f"{'Subject':<15} {'Marks':<10} {'Grade':<5}")

for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]
    grade = get_grade(mark)
    print(f"{subject:<15} {mark:<10} {grade:<5}")


# print and calculate these
total_marks = sum(marks)
average_marks = round(total_marks / len(marks), 2)
highest_mark = max(marks)
lowest_mark = min(marks)
highest_subject = subjects[marks.index(highest_mark)]
lowest_subject = subjects[marks.index(lowest_mark)]

print(f"Total Marks:           {total_marks}")
print(f"Average Marks:         {average_marks}")
print(f"Highest Scoring:       {highest_subject} ({highest_mark})")
print(f"Lowest Scoring:        {lowest_subject} ({lowest_mark})")


# while loop for adding new subjects ---
print("\n--- Add New Subjects ---")
print("Type 'done' as subject name to finish.\n")

new_subjects_count = 0

while True:
    subject_input = input("Enter subject name: ").strip()

    # Check for exit condition
    if subject_input.lower() == "done":
        break

    # Validate subject name not empty
    if not subject_input:
        print("do not crash, and do not add invalid entries to the list.\n")
        continue

    # Get marks input
    marks_input = input(f"Enter marks for {subject_input} (0-100): ").strip()

    # Validate marks are integer
    try:
        mark_value = float(marks_input)

        # Check range
        if mark_value < 0 or mark_value > 100:
            print("do not crash, and do not add invalid entries to the list.\n")
            continue

        # Convert to int if a whole number
        if mark_value == int(mark_value):
            mark_value = int(mark_value)

        # Add valid entry
        subjects.append(subject_input)
        marks.append(mark_value)
        new_subjects_count += 1
        print(f"✓ Added: {subject_input} - {mark_value} ({get_grade(mark_value)})\n")

    except ValueError:
        print(" do not crash, and do not add invalid entries to the list.\n")


# Final Summary
print(f"New subjects added:    {new_subjects_count}")

if len(marks) > 0:
    updated_average = round(sum(marks) / len(marks), 2)
    print(f"Updated average:       {updated_average}")
    print(f"Total subjects now:    {len(subjects)}")

if new_subjects_count > 0:
    print("\nUpdated subject list:")
    for i in range(len(subjects)):
        print(f"  • {subjects[i]}: {marks[i]} ({get_grade(marks[i])})")
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# check for each student
results = []
for name, scores in class_data:
    avg = round(sum(scores) / len(scores), 2)
    status = "Pass" if avg >= 60 else "Fail"
    results.append((name, avg, status))

# Print header
print(f"{'Name':<18}| {'Average':^7} | Status")
print("-" * 40)

# Print each student's row
for name, avg, status in results:
    print(f"{name:<18}| {avg:^7.2f} | {status}")

# Calculate pass fail topper and avg
passed = sum(1 for _, _, status in results if status == "Pass")
failed = len(results) - passed
topper = max(results, key=lambda x: x[1])
class_avg = round(sum(avg for _, avg, _ in results) / len(results), 2)

# Print summaary
print()
print(f"Passed: {passed}")
print(f"Failed: {failed}")
print(f"Class Topper: {topper[0]} ({topper[1]})")
print(f"Class Average: {class_avg}")
```
